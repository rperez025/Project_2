```python
# Import the dependencies
import pandas as pd
```

```python
# Set the column width
pd.options.display.max_colwidth = 200

# Read in the first dataset. 
foodWasteDF = pd.read_csv('Food Waste data and research - by country.csv')
foodWasteDF.head()
```

```python
# Get the index datatype (column names) for the first dataset  (NOTE: 214 Countries noted).
foodWasteDF.info()
```

```python
# Read in the second dataset. 
globalSalaryDF = pd.read_csv('salary_data.csv')
globalSalaryDF.head()
```

```python
# Get the index datatype (column names) for the second dataset (NOTE: 221 Countries noted).
globalSalaryDF.info()
```

```python
# Rename the columns. 
globalSalaryDF = globalSalaryDF.rename(columns = {
    'country_name': 'Country',
    'average_salary': 'Avg Salary'
})

globalSalaryDF.head()
```

```python
# Identified countries to be renamed
# Convert the 'Country' columns in both DataFrames to sets
countries_in_foodWasteDF = set(foodWasteDF['Country'])
countries_in_globalSalaryDF = set(globalSalaryDF['Country'])

# Find countries unique to foodWasteDF
unique_to_foodWasteDF = countries_in_foodWasteDF - countries_in_globalSalaryDF

# Find countries unique to globalSalaryDF
unique_to_globalSalaryDF = countries_in_globalSalaryDF - countries_in_foodWasteDF

# Find countries common to both DataFrames
common_countries = countries_in_foodWasteDF & countries_in_globalSalaryDF

# Print the results
#print("Unique to foodWasteDF:")
#print(sorted(list(unique_to_foodWasteDF)))
#print("\n-------------------------\n")  # Separator with line breaks

#print("Unique to globalSalaryDF:")
#print(sorted(list(unique_to_globalSalaryDF)))
#print("\n-------------------------\n")  # Separator with line breaks

#print("Common countries:")
#print(sorted(list(common_countries)))
```

```python
# Need to update the names of countries

rename_dict = {'Bolivia (Plurin. State of)': 'Bolivia',
               'Brunei Darussalam': 'Brunei',
               'Cabo Verde': 'Cape Verde',
               'China, Hong Kong SAR': 'Hong Kong',
               'China, Macao SAR': 'Macao',
               'Czechia': 'Czech Republic',
               "Dem. People's Rep. Korea": 'Korea (North)',
               'Dem. Rep. of the Congo': 'Congo Democratic Republic',
               'Eswatini': 'Swaziland',
               'Iran (Islamic Republic of)': 'Iran',
               "Lao People's Dem. Rep.": "Laos",
               'Micronesia (Fed. States of)': 'Micronesia',
               'North Macedonia': 'Macedonia',
               'Republic of Korea': 'Korea (South)',
               'Republic of Moldova': 'Moldova',
               'Russian Federation': 'Russia',
               'State of Palestine': 'Palestine',
               'United Rep. of Tanzania': 'Tanzania',
               'Saint Vincent & Grenadines': 'Saint Vincent and the Grenadines',
               'Syrian Arab Republic': 'Syria',
               'Timor-Leste': 'East Timor',
               'United States Virgin Islands': 'Virgin Islands (US)',
               'United States of America': 'United States',
               'Venezuela (Boliv. Rep. of)': 'Venezuela',
               'Viet Nam': 'Vietnam',
               }

# Use replace method to rename 
foodWasteDF['Country'] = foodWasteDF['Country'].replace(rename_dict)

# Updated dataframe
foodWasteDF.head()
```

```python
# Combine both DataFrames using outer join to obtain all countries from both DataFrames
combinedDF = pd.merge(foodWasteDF, globalSalaryDF, on = 'Country', how = 'left')
combinedDF.head(5)
```

```python
from pprint import pprint
```

```python
# Convert the dataframe to a list of dictionaries to be prepared to import into MongoDB
foodWaste_DICT = list(combinedDF.to_dict('records'))
pprint(foodWaste_DICT)
```

```python
from pymongo import MongoClient
```

```python
# instance of MonogClient
mongo = MongoClient(port=27017)
```

```python
# reference to a new database & its collection to create it
db = mongo.Project2_DB
collection = db.Project2_collection
```

```python
# use insert_many() to insert data from the extracted DataFrame info # misnomer, I actually went back up & made it a list
collection.insert_many(foodWaste_DICT)
```

```python
# check that it was made by piping it into a results query, with find() & a for look with a print statement
query = collection.find_one()
print(query)
```
